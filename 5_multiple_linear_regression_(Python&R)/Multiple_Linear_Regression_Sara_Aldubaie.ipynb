```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV

from sklearn.datasets import load_boston
```

```python
boston = load_boston()

# Boston Housing Dataset
print(boston['DESCR'])
```

```python
# load keys from boston dict
boston.keys()
```

```python
# put it into a pandas dataframe

df = pd.DataFrame(boston['data'])

df
```

```python
boston['feature_names']
```

```python
# set column names to feature names
df.columns = boston['feature_names']

df
```

```python
# add in the price (target) column
df['target'] = boston['target']

df
```

```python
# to save the dataset, and be able use it in R 
df.to_csv(r'.\boston.csv')
```

```python
corr = df.corr()

corr.style.background_gradient(cmap='coolwarm')
```

## <span style="color:#EF7C8E">First Model</span>

```python

# feature = ZN , RM , DIS, B. since all of them have positve corrlatiom with target 

vars = [
        'ZN' , 'RM' , 'DIS', 'B'
        ]

X = df[vars]
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    train_size=0.75, 
                                                    random_state=42
                                                   )

multi_lr = LinearRegression()

multi_lr.fit(X_train, y_train)

multi_predictions = multi_lr.predict(X_test)


r2_score_1 = r2_score(y_true=y_test, y_pred=multi_predictions)
```

## <span style="color:#EF7C8E">Second Model</span>

```python
# feature = all columns

vars = [
        'CRIM', 'ZN' ,'INDUS' ,'CHAS', 'NOX' ,'RM' ,'AGE' ,'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'
        ]

X = df[vars]
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    train_size=0.75, 
                                                    random_state=42
                                                   )

multi_lr = LinearRegression()

multi_lr.fit(X_train, y_train)

multi_predictions = multi_lr.predict(X_test)


r2_score_2 = r2_score(y_true=y_test, y_pred=multi_predictions)
```

## <span style="color:#EF7C8E">Third Model</span>

```python
# feature = all columns but without 'INDUS'

vars = [
        'NOX', 'RM', 'DIS', 'RAD', 'PTRATIO', 'LSTAT', 'CRIM', 'ZN', 'CHAS' , 'TAX', 'B'
        ]

X = df[vars]
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    train_size=0.75, 
                                                    random_state=42
                                                   )

multi_lr = LinearRegression()

multi_lr.fit(X_train, y_train)

multi_predictions = multi_lr.predict(X_test)


r2_score_3 = r2_score(y_true=y_test, y_pred=multi_predictions)
```

## <span style="color:#EF7C8E">Forth Model</span>

```python
# feature = most negative correlation and most positive correlation  


vars = [
        'RM','LSTAT'
        ]

X = df[vars]
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    train_size=0.75, 
                                                    random_state=42
                                                   )

multi_lr = LinearRegression()

multi_lr.fit(X_train, y_train)

multi_predictions = multi_lr.predict(X_test)


r2_score_4 = r2_score(y_true=y_test, y_pred=multi_predictions)
```

```python

print(f"First Model R2 score: {r2_score_1}")
print(f"Second Model R2 score: {r2_score_2}")
print(f"Third Model R2 score: {r2_score_3}")
print(f"Forth Model R2 score: {r2_score_4}")
```

## <span style="color:#EF7C8E">Result:  </span>
